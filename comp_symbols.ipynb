```python
import pandas as pd
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
url = "https://mse.mn/mn/trade_today/23"
html = urlopen(url)
soup = BeautifulSoup(html)
```

```python
soup
```

```python
soup.find_all("option")[1:]
```

```python
data = {}
for row in soup.find_all("option")[1:]:
    c, s = row.text.split(' | ')
    data.setdefault('Компаний нэр', []).append(c)
    data.setdefault('Символ', []).append(s)
    data.setdefault('Индекс', []).append(row['value'])
pd.DataFrame(data).to_csv('Data/comp_symbols.csv', index=False)
print(data)
```
